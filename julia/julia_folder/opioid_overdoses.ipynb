```python
for state in unique_states:
    state_2016 = overdoses_2016_df.loc[overdoses_2016_df["state_name"] == state, :]
    state_2016_df = state_2016.groupby("indicator").sum()
    state_2016_df.rename(columns={"data_value": state})
    print(state)
    print(state_2016_df)
    print("--------------------")
```

```python
state_dicts = []
```

```python
for state in unique_states:
    
    State_2016 = overdoses_2016_df.loc[overdoses_2016_df["state"] == state, :]
    State_2016_df = State_2016.groupby("indicator").sum()
    state_dict = State_2016_df.to_dict()
    state_final_dict = state_dict['data_value'] 
    state_final_dict["State"] = state
    state_dicts.append(state_final_dict)
```

```python
state_dicts
```

```python
total_df = pd.DataFrame(state_dicts)

total_df.head()
```

```python
renamed = total_df.rename(columns={'Number of Deaths': 'Total Number of Deaths', 'Opioids (T40.0-T40.4,T40.6)': 'Opioid Overdose Deaths', 'Heroin (T40.1)': 'Heroin Overdose Deaths', 'Natural & semi-synthetic opioids (T40.2)': 'Natural & Semi-synthetic Opioid Overdose Deaths', 'Synthetic opioids, excl. methadone (T40.4)': 'Synthetic Opioid Deaths (excl. methadone)', 'Methadone (T40.3)': 'Methadone Overdose Deaths'})
                         
renamed.head()
```

```python
renamed['% of Drug Overdose Deaths'] = (renamed['Number of Drug Overdose Deaths']/renamed['Total Number of Deaths'])*100
renamed = renamed[renamed.State != 'US']
renamed.head()
```

```python
renamed_df = renamed[['State', 'Total Number of Deaths', 'Number of Drug Overdose Deaths', '% of Drug Overdose Deaths', 'Opioid Overdose Deaths', 'Heroin Overdose Deaths', 'Natural & Semi-synthetic Opioid Overdose Deaths', 'Synthetic Opioid Deaths (excl. methadone)', 'Methadone Overdose Deaths']]

Opioid_overdose_deaths_2016 = renamed_df.sort_values(by='Opioid Overdose Deaths', ascending=False)

Opioid_overdose_deaths_2016
```
